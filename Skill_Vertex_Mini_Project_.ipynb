# CLEANING THE DATA

```python
#Gives the sum of null values
data.isnull().sum()
```

```python
#White spot in the figure represent presense of null values
plt.figure(figsize=(25,25))
sns.heatmap(data.isnull())
```

```python
#Prints percentage of null values in particular variable or column
#axis 0(zero)=gives rows
null_var=data.isnull().sum()/data.shape[0]*100
null_var
```

```python
#Gives the names of columns what has more that 8% null values in it
drop_columns=null_var[null_var>8].keys()
drop_columns
```

```python
#Dropping columns which has null values in it
df2_drop_clm=data.drop(columns=drop_columns)
```

```python
#Gives shape of dataset after dropping the columns 
df2_drop_clm.shape
```

```python
sns.heatmap(df2_drop_clm.isnull())
```

```python
#dropna function = Drops the rows which has null values in it
df3_drop_rows=df2_drop_clm.dropna()
```

```python
df3_drop_rows.shape
```

```python
sns.heatmap(df3_drop_rows.isnull())
```

DATASET HAS NO NULL VALUES IN IT NOW

```python
#Checks if there is any null value present in dataset
df3_drop_rows.isnull().sum().sum()
```

There is no more null values in our dataset now

```python
#Distribution of new dataset which has no null values in it
#displot only plots numeric data
sns.distplot(df3_drop_rows['Age'])
```

```python
#Overlapping original and new dataset's distplot of Age column
plt.figure(figsize=(25,15))
sns.distplot(data['Age'])
sns.distplot(df3_drop_rows['Age'])
```

```python
#Assigning all numerical columns to a variable
num_var=[ 'Age', 'Overall', 'Potential', 'Special',
       'International Reputation', 'Weak Foot', 'Skill Moves', 'Jersey Number',
       'Crossing', 'Finishing', 'HeadingAccuracy', 'ShortPassing', 'Volleys',
       'Dribbling', 'Curve', 'FKAccuracy', 'LongPassing', 'BallControl',
       'Acceleration', 'SprintSpeed', 'Agility', 'Reactions', 'Balance',
       'ShotPower', 'Jumping', 'Stamina', 'Strength', 'LongShots',
       'Aggression', 'Interceptions', 'Positioning', 'Vision', 'Penalties',
       'Composure', 'Marking', 'StandingTackle', 'SlidingTackle', 'GKDiving',
       'GKHandling', 'GKKicking', 'GKPositioning', 'GKReflexes']
print(len(num_var))
```

```python
#Data Distribution of all numeric columns
plt.figure(figsize=(25,25))
#enumerate function gives an item from list with its index
for i,var in enumerate(num_var):
    #There are 42 columns so 7 multiplied to 6 is equal to 42
    plt.subplot(7,6,i+1)
    sns.distplot(data[var],bins=20)
    sns.distplot(df3_drop_rows[var],bins=20)
    
```

```python
#Gives the names of categorical columns 
df3_drop_rows.select_dtypes(include=['object']).columns
```

```python
cat_val=['Name', 'Photo', 'Nationality', 'Flag', 'Club', 'Club Logo', 'Value',
       'Wage', 'Preferred Foot', 'Work Rate', 'Body Type', 'Real Face',
       'Position', 'Contract Valid Until', 'Height', 'Weight']
print(len(cat_val))
```

```python
#Calculating percentage of Body Type column of original dataset = (value/rows)*100 
#axis 0(zero) means rows
print(data['Body Type'].value_counts()/data.shape[0]*100)
print('\n')
#Calculating percentage of Body Type column of new dataset = (value/rows)*100 
print(df3_drop_rows['Body Type'].value_counts()/df3_drop_rows.shape[0]*100)
```

```python
#Distribution of Categorical Data of Body Type Column
pd.concat([data['Body Type'].value_counts()/data.shape[0]*100,
         df3_drop_rows['Body Type'].value_counts()/df3_drop_rows.shape[0]*100],axis=1,
         keys=['Body Type_org','Body Type_clean'])
```

```python
#Defining a function to show distribution of all categorical data columns
def cat_var_dist(var):
    return pd.concat([data[var].value_counts()/data.shape[0]*100,
           df3_drop_rows[var].value_counts()/df3_drop_rows.shape[0]*100],axis=1,
           keys=[var+'_org',var+'_clean'])  
    
```

```python
cat_var_dist('Club Logo')
```

```python
#Distribution of all categorical columns
for i in range(0,len(cat_val),1):
    def cat_var_dist(var):
          return pd.concat([data[var].value_counts()/data.shape[0]*100,
                 df3_drop_rows[var].value_counts()/df3_drop_rows.shape[0]*100],axis=1,
                 keys=[var+'_org',var+'_clean'])
    print('\n\n')
    print("Categorical column = ",i+1)
    print('\n')
    print(cat_var_dist(cat_val[i]))
```
